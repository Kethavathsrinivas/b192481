```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s={"iris species":33.3,"Iris vergina":33.3,"iris versicolor":33.3}
```

```python
c=list(s.keys())
p=list(s.values())
print(c)
print(s)
```

```python
fig=plt.figure(figsize=(10,6))
plt.pie(x=p,labels=c,explode=[0.1,0.1,0.1],colors=["blue","orange","green"],shadow=True,autopct="%0.2f%%")
plt.legend(["Iris species","Iris vergina","iris versicolor"])
plt.title("Iris species %")
plt.ylabel("species")
plt.show()
```
